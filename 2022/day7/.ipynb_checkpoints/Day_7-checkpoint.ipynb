```python
def foldersizes(filedict):
    folder = {}
    
    for filename, sizefile in filedict.items():
        sumsize = 0
        #print(filename)
        path = filename.split('/')[:-1]
        path = '/'.join(path)
        
        # size of all files in one folder
        if path in folder:
            # is already an folder
            sumsize = folder[path]
            sumsize += int(sizefile)
            folder[path]=sumsize
        else:
            # is an new folder
            sumsize += int(sizefile)
            folder[path]=sumsize
        
        # sum up all subfolders
        for foldername, foldersize in folder.items():
            if path in foldername and path != foldername:
                sumsize += foldersize
                folder[foldername] = sumsize      
    
    #for foldername, foldersize in folder.items():
    #    print(f'Folder {foldername} hat {foldersize}')
    
    return folder
        
x = foldersizes(x)
```

```python
x_list = sorted(list(x.items()), key = lambda key : len(key[0]))

x = {ele[0] : ele[1]  for ele in x_list}
```
