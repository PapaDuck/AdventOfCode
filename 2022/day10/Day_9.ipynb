```python
def cycles(cyclelist):
    cycleno = 1
    x = 1
    sumsignalstrengths = 0
    for cycle in cyclelist:
        # START OF CYCLE
        cycleno += 1
        if cycle[:4] == 'addx':
            _, value = cycle.split(' ')
            value = int(value)
        elif cycle[:4] == 'noop':
            value = 0
        
        # 20th, 60th cycle ...
        if cycleno == 20 or (cycleno+20) in range(0,1000000,40):
            print(str(cycleno)+' '+cycle+' --- > '+str(x)+' < ---')
            strength = cycleno*x
            print('-----'+str(strength))
            sumsignalstrengths += strength
        else:
           print(str(cycleno)+' '+cycle+' '+str(x))
        
        # END OF CYCLE
        x += value
        
        cycleno += 1
    return sumsignalstrengths
cycles(input)
```
